# Naver NLP Challenge -> klue/ner 형태로

```python
fin_naver = open("ner_dataset/naver-nlp-challenge/train_data", "r")
lines = fin_naver.readlines()
len(lines)
```

```python
sentence_list = []
tokens_list = []
ner_tags_list = []

words = []
tokens = []
ner_tags = []
for line in lines:
    if line == '\n':
        sentence = ' '.join(words)
        sentence_list.append(sentence)
        tokens_list.append(tokens)
        ner_tags_list.append(ner_tags)
        words, tokens, ner_tags = [], [], []
    else:
        seps = line.split('\t')
        words.append(seps[1])
        token = list(seps[1])
        tokens.extend(token)
        ner_tags.extend([seps[2][:-1]]*len(token))
        
datadict = {
    'sentence': sentence_list.copy(),
    'tokens': tokens_list.copy(),
    'ner_tags': ner_tags_list.copy(),
}
```

```python
print(len(datadict['sentence']))

for i in range(10000, 10010):
    print(datadict['sentence'][i])
    print(datadict['tokens'][i])
    print(datadict['ner_tags'][i])
```

```python
# 토큰 만드는 건 연구실에 있는 코드 써서 하자.
# 일단 어떤 것들이 나오는지 한번 봅시다.
tags_stat = {}
for sentence in sentences_list:
    seps = sentence.split(':')[1:]
    for sep in seps:
        tag = sep.split('>')[0]
        if tag in tags_stat:
            tags_stat[tag] += 1
        else:
            tags_stat[tag] = 1
            
tags_stat
```

```python
def get_sents(lines):
    sentences_list = []
    for line in lines:
        if line.startswith('$'): sentences_list.append(line[1:-1])

    return sentences_list
```

```python
fin_kmou_dev = open("ner_dataset/kmounlp/ner.dev", "r")
lines_kmou_dev = fin_kmou_dev.readlines()
fin_kmou_dev.close()
print(lines_kmou_dev.count('\n'))

sents_kmou_dev = get_sents(lines_kmou_dev)
print(sents_kmou_dev[:3])
```

```python
fin_kmou_test = open("ner_dataset/kmounlp/ner.test", "r")
lines_kmou_test = fin_kmou_test.readlines()
fin_kmou_test.close()
print(lines_kmou_test.count('\n'))

sents_kmou_test = get_sents(lines_kmou_test)
print(sents_kmou_test[:3])
```

```python
fin_kmou_train = open("ner_dataset/kmounlp/ner.train", "r")
lines_kmou_train = fin_kmou_train.readlines()
fin_kmou_train.close()
print(lines_kmou_train.count('\n'))

sents_kmou_train = get_sents(lines_kmou_train)
print(sents_kmou_train[:3])
```

```python
def tag_counter(sents):
    tags_stat = {}
    for sentence in sents:
        seps = sentence.split(':')[1:]
        for sep in seps:
            tag = sep.split('>')[0]
            if tag in tags_stat:
                tags_stat[tag] += 1
            else:
                tags_stat[tag] = 1

    return tags_stat
```

```python
print("dev:",tag_counter(sents_kmou_dev))
print("test:",tag_counter(sents_kmou_test))
print("train:",tag_counter(sents_kmou_train))
```
