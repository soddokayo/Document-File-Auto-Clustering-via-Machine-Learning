```python
from transformers import pipeline
```

```python
classifier = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')
```

```python
labels = ["artifacts", "animals", "food", "birds"]
hypothesis_template = 'This text is about {}.'
sequence = "I like hot dogs"
```

```python
prediction = classifier(sequence, labels, hypothesis_template=hypothesis_template, multi_class=True)
print(prediction)
```

```python
# AG_NEWS로 테스트
from torchtext.datasets import AG_NEWS

test_iter = AG_NEWS(split='test')

#def tokenize(label, line):
#    return line.split()

labels = []
lines = []
for label, line in test_iter:
    labels.append(label)
    lines.append(line)
    
for i in range(25, 35):
    print("["+str(i)+"]", labels[i], lines[i])
```

```python
labels_agnews = ["World", "Sports", "Business", "Sci/Tech"]
hypothesis_template = 'This text is about {}.'
sequences = lines.copy()
answers = labels.copy()

print("predicting", len(lines), "records of test dataset...")
cnt = 0
for idx, sequence in enumerate(sequences):
    print(idx, end='] ')#sequence, answers[idx])
    prediction = classifier(sequence, labels_agnews, hypothesis_template=hypothesis_template, multi_label=True)
    print("pred: "+prediction['labels'][0], "/", "answ: "+labels_agnews[answers[idx]-1])

    if prediction['labels'][0] == labels_agnews[answers[idx]-1]:
        cnt += 1
        
print("results:", cnt, "of", len(lines), "correct!")
print("ACC: ", cnt/len(lines))
```
