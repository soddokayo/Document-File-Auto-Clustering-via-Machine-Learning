# AGNEWS

```python
from transformers import pipeline

classifier = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')
```

```python
# test
labels = ["artifacts", "animals", "food", "birds"]
hypothesis_template = 'This text is about {}.'
sequence = "I like hot dogs"

prediction = classifier(sequence, labels, hypothesis_template=hypothesis_template, multi_class=True)
print(prediction)
```

```python
# AG_NEWS로 테스트
from torchtext.datasets import AG_NEWS

test_iter = AG_NEWS(split='test')

#def tokenize(label, line):
#    return line.split()

labels = []
lines = []
for label, line in test_iter:
    labels.append(label)
    lines.append(line)
    
for i in range(25, 27):
    print("["+str(i)+"]", labels[i], lines[i])
```

```python
# tqdm test
from tqdm import tqdm
import time
 
for i in tqdm(range(10)): ## tqdm 첫 번째 인자에 순회가능한 객체 전달
    time.sleep(0.1)
```

```python
from tqdm import tqdm

labels_agnews = ["world", "sports", "business", "science and technology"]
hypothesis_template = 'This text is about {}.'
sequences = lines.copy()
answers = labels.copy()

print("predicting", len(lines), "records of test dataset...")
cnt = 0
preds = []
progressbar = tqdm(range(len(lines)))
for idx in progressbar:
    #print(idx, end='] ')#sequence, answers[idx])
    prediction = classifier(sequences[idx], labels_agnews, hypothesis_template=hypothesis_template, multi_label=True)
    #print("pred: "+prediction['labels'][0], "/", "answ: "+labels_agnews[answers[idx]-1])

    preds.append(prediction['labels'][0])
    if prediction['labels'][0] == labels_agnews[answers[idx]-1]:
        cnt += 1

progressbar.close()
print("results:", cnt, "of", len(lines), "correct!")
print("ACC: ", cnt/len(lines))
```

```python
# preds 저장
import pickle, gzip

# save and compress.
with gzip.open('preds_agnews.pickle', 'wb') as f:
    pickle.dump(preds, f)

# load and uncompress.
with gzip.open('preds_agnews.pickle','rb') as f:
    preds_agnews = pickle.load(f)
```

# IMDB

```python
from transformers import pipeline

classifier = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')

# test
labels = ["artifacts", "animals", "food", "birds"]
hypothesis_template = 'This text is about {}.'
sequence = "I like hot dogs"

prediction = classifier(sequence, labels, hypothesis_template=hypothesis_template, multi_class=True)
print(prediction)
```

```python
# IMDB로 테스트
from torchtext.datasets import IMDB

test_iter_imdb = IMDB(split='test')

labels_imdb = []
lines_imdb = []
for label, line in test_iter_imdb:
    labels_imdb.append(label)
    lines_imdb.append(line)
    
for i in range(25, 27):
    print("["+str(i)+"]", labels_imdb[i], lines_imdb[i])
```

```python
print(len(labels_imdb), len(lines_imdb))
```

```python
# sentence parsing test
from nltk.tokenize import sent_tokenize
    
for i in range(25, 27):
    tokens = sent_tokenize(lines_imdb[i])
    print("["+str(i)+"]", labels_imdb[i], len(tokens), tokens)
```

```python
from tqdm.notebook import tqdm

cats_imdb_det = ["positive", "negative"]
#hypothesis_template = 'This text is about {}.'
sequences = lines_imdb[21000:].copy()
answers = labels_imdb[21000:].copy()

print("predicting", len(answers), "records of test dataset...")
cnt = 0
preds_imdb = []
for i in tqdm(range(len(answers)), desc='outer', position=0):
    #print("["+str(i)+"]", end='')
    tokens = sent_tokenize(sequences[i])
    vote = []
    for j in tqdm(range(len(tokens)), desc='inner', position=1, leave=False):
        prediction = classifier(tokens[j], cats_imdb_det)
        #print(prediction)
        vote.append(prediction['labels'][0])

    preds_imdb.append(max(vote, key=vote.count))
    #print("vote: ", preds_imdb[i][:3], "(", vote.count('positive'), vote.count('negative'), ") / answ: ", answers[i])
    if preds_imdb[i][:3] == answers[i]:
        #print("correct!")
        cnt += 1

print("results:", cnt, "of", len(answers), "correct!")
print("ACC: ", cnt/len(answers))
```
