## GoogleTrans

```python
from googletrans import Translator

tr = Translator()

cats_ynat = []
for cat in cats_ynat_ko:
    translated = tr.translate(cat)
    print(translated)
    cats_ynat.append(translated.text)
    
print(cats_ynat)
```

```python
cats_ynat = ['IT or science', 'economy', 'society', 'life and culture', 'world', 'sports', 'politics']
print(cats_ynat)
```

```python
from tqdm.notebook import tqdm

lines_ynat_ko = lines_ynat.copy()
lines_ynat_en = []
for i in tqdm(range(len(lines_ynat_ko))):
    translated = tr.translate(lines_ynat_ko[i])
    print(lines_ynat_ko[i])
    print("-> ", translated.text)
    lines_ynat_en.append(translated.text)
    
print(lines_ynat_en)
```

```python
# 너무 많이 번역해서 밴당함 ㅋ
print(len(lines_ynat_en))
for i in range(len(lines_ynat_en)):
    print(lines_ynat_ko[i])
    print("-> ", lines_ynat_en[i])
```

```python
# 1651개 번역된거 일단 저장
import pickle, gzip

# save and compress.
with gzip.open('lines_ynat_en[:1651].pickle', 'wb') as f:
    pickle.dump(lines_ynat_en, f)

# load and uncompress.
with gzip.open('lines_ynat_en[:1651].pickle','rb') as f:
    lines_ynat_reloaded = pickle.load(f)
    
print(len(lines_ynat_reloaded), lines_ynat_reloaded)
```

## Zero Shot Classification

```python
from transformers import pipeline

classifier = pipeline(model='facebook/bart-large-mnli')

# test
labels = ["artifacts", "animals", "food", "birds"]
#hypothesis_template = 'This text is about {}.'
sequence = "I like hot dogs"

prediction = classifier(sequence, labels)
print(prediction)
```

```python
# 다시 불러와
import pickle, gzip

# load and uncompress.
with gzip.open('lines_ynat_en[:1651].pickle','rb') as f:
    lines_ynat_reloaded = pickle.load(f)
    
print(len(lines_ynat_reloaded), lines_ynat_reloaded)
```

```python
cats_ynat_det = [
    'IT or science', #0
    'economy', #1
    'society', #2
    'life and culture', #3
    'world', #4
    'sports', #5
    'politics' #6
]
#hypothesis_template = 'This text is about {}.'
sequences = lines_ynat_reloaded.copy()
answers = labels_ynat.copy()

print(len(answers), min(answers), max(answers))
```

```python
from tqdm.notebook import tqdm

print("predicting", len(sequences), "records of test dataset...")
cnt = 0
preds_ynat = []
for i in tqdm(range(len(sequences))):
    print("["+str(i)+"]", end='')
    prediction = classifier(sequences[i], cats_ynat_det)
    #print(prediction)
    preds_ynat.append(prediction['labels'][0])
    
    print("pred: ", preds_ynat[i], "/ answ: ", cats_ynat_det[answers[i]])
    if preds_ynat[i] == cats_ynat_det[answers[i]]:
        print("correct!")
        cnt += 1

print("results:", cnt, "of", len(sequences), "correct!")
print("ACC: ", cnt/len(sequences))
```
