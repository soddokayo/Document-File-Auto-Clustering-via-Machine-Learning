```python
# preds 저장
import pickle, gzip

# load and uncompress.
with gzip.open('lines_ynat_en[:1651].pickle','rb') as f:
    lines_ynat_en_reloaded_1 = pickle.load(f)

# load and uncompress.
with gzip.open('lines_ynat_en[1500:3000].pickle','rb') as f:
    lines_ynat_en_reloaded_2 = pickle.load(f)
    
print(len(lines_ynat_en_reloaded_1[:1500]), len(lines_ynat_en_reloaded_2))
print(len(lines_ynat_en_reloaded_1[:1500])+len(lines_ynat_en_reloaded_2))
```

```python
lines_ynat_en = lines_ynat_en_reloaded_1[:1500].copy() + lines_ynat_en_reloaded_2.copy()
print(len(lines_ynat_en), lines_ynat_en[1500-10:1500+10])
```

```python
from datasets import load_dataset
dataset = load_dataset('klue', 'ynat')

labels_ynat = []
lines_ynat = []
for i in range(int(dataset['validation'].num_rows)): # 9107개
    label = dataset['validation'][i]['label']
    line = dataset['validation'][i]['title']
    labels_ynat.append(label)
    lines_ynat.append(line)
    
for i in range(1125, 1135):
    print("["+str(i)+"]", labels_ynat[i], lines_ynat[i])
```

```python
cats_ynat = ['IT or science', 'economy', 'society', 'life and culture', 'world', 'sports', 'politics']
print(cats_ynat)
```

### Zero-shot

```python
from transformers import pipeline

classifier = pipeline(model='facebook/bart-large-mnli')

# test
labels = ["artifacts", "animals", "food", "birds"]
#hypothesis_template = 'This text is about {}.'
sequence = "I like hot dogs"

prediction = classifier(sequence, labels)
print(prediction)
```

```python
cats_ynat_det = [
    'IT or science', #0
    'economy', #1
#    'society', #2
#    'life and culture', #3
#    'world', #4
#    'sports', #5
#    'politics' #6
]
#hypothesis_template = 'This text is about {}.'
sequences = lines_ynat_en.copy()
answers = labels_ynat.copy()

print(len(answers), min(answers), max(answers))
```

```python
from tqdm.notebook import tqdm

print("predicting", len(sequences), "records of test dataset...")
tot = 0
cnt = 0
idx_01 = []
preds_ynat = []

# for computing F1-Score
preds = []
answ = []
for i in tqdm(range(len(sequences))):
    if answers[i] not in [0, 1]:
        preds_ynat.append('')
        continue
    tot += 1
    idx_01.append(i)
    print("["+str(i)+"]", end='')
    prediction = classifier(sequences[i], cats_ynat_det)
    #print(prediction)
    preds_ynat.append(prediction['labels'][0])
    
    # for computing F1-Score
    print("pred: ", preds_ynat[i], "/ answ: ", cats_ynat_det[answers[i]])
    print("ko: ", lines_ynat[i])
    print("en: ", sequences[i])
    if preds_ynat[i] == cats_ynat_det[answers[i]]:
        print("correct!")
        cnt += 1
        
    preds.append(preds_ynat[i])
    answ.append(cats_ynat_det[answers[i]])

print("results:", cnt, "of", tot, "correct!")
print("ACC: ", cnt/tot)
```

```python
preds2 = []
answ2 = []
for i in range(len(sequences)):
    if i in idx_01:
        preds2.append(preds_ynat[i])
        answ2.append(cats_ynat_det[answers[i]])
        
if preds == preds2:
    print("preds verified")
if answ == answ2:
    print("answ verified")
    
print(len(preds2), len(answ2))
```

```python
print("results:", cnt, "of", tot, "correct!")
print("ACC: ", cnt/tot)
```

```python
# true: 0, false: 1이라 가정하면
true_positive = 0
false_negative = 0
false_positive = 0
true_negative = 0
for i in range(len(preds2)):
    if preds[i]==cats_ynat_det[0] and answ[i]==cats_ynat_det[0]:
        true_positive += 1
    elif preds[i]==cats_ynat_det[1] and answ[i]==cats_ynat_det[0]:
        false_negative += 1
    elif preds[i]==cats_ynat_det[0] and answ[i]==cats_ynat_det[1]:
        false_positive += 1
    elif preds[i]==cats_ynat_det[1] and answ[i]==cats_ynat_det[1]:
        true_negative += 1

print(true_positive, false_negative, false_positive, true_negative)
recall = true_positive/(true_positive + false_negative)
precision = true_positive/(true_positive + false_positive)
f1_score = 2*precision*recall/(precision + recall)
print(recall, precision, f1_score)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

print(accuracy_score(answ, preds))
print(recall_score(answ, preds, pos_label=cats_ynat_det[0]))
print(precision_score(answ, preds, pos_label=cats_ynat_det[0]))
print(f1_score(answ, preds, pos_label=cats_ynat_det[0]))
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

print(accuracy_score(answ, preds))
print(recall_score(answ, preds, pos_label=cats_ynat_det[1]))
print(precision_score(answ, preds, pos_label=cats_ynat_det[1]))
print(f1_score(answ, preds, pos_label=cats_ynat_det[1]))
```
