```python
#from sklearn.feature_extraction.text import TfidfVectorizer
import pickle

tfidf_voca_agnews = pickle.load(open("tfidf_voca_agnews.pkl", "rb"))
print("vocabulary: ", len(tfidf_voca_agnews), tfidf_voca_agnews)

tfidf_arr_agnews = pickle.load(open("tfidf_arr_agnews.pkl", "rb"))
print("array: ", tfidf_arr_agnews.shape[0], tfidf_arr_agnews.shape[1], tfidf_arr_agnews)
```

```python
print(type(tfidf_voca_agnews), type(tfidf_arr_agnews))
```

```python
#      https://scikit-learn.org/stable/auto_examples/text/plot_document_clustering.html#clustering-sparse-data-with-k-means

import numpy as np
from sklearn.cluster import KMeans

# kmeans = [0,0,0,0,0,0,0]
for seed in range(1,3):
    kmeans[seed] = KMeans(
        n_clusters=4,  # AGNEWS 4!
        max_iter=100,
        n_init=5,
        random_state=seed,
    ).fit(tfidf_arr_agnews)
    cluster_idf, cluster_sizes = np.unique(kmeans[seed].labels_, return_counts=True)
    print(f"# of elements asigned to each cluster: {cluster_sizes}")
```

```python
print(kmeans)
```

```python
print(kmeans[4].labels_)

cluster_idf, cluster_sizes = np.unique(kmeans[4].labels_, return_counts=True)
print(cluster_idf, cluster_sizes)
```

```python
original_space_centroids = kmeans[4].cluster_centers_
order_centroids = original_space_centroids.argsort()[:, ::-1]
terms = tfidf_voca_agnews

for i in range(4):
    print(f"Cluster {i}: ", end="")
    for ind in order_centroids[i, :10]:
        word = [key for key, value in terms.items() if value == ind]
        print(f"{word[0]} ", end="")
    print()
```

```python
for i in range(4):
    print(f"Cluster {i}: ", end="")
    for ind in order_centroids[i, :10]:
        print(f"{ind} ", end="")
    print()
```
